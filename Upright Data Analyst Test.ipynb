```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales=np.genfromtxt(r'C:\Users\surfacepro\Downloads\data.csv',delimiter=',',skip_header=1)
print(sales)
```

```python
months=data.readline().strip().split(',')
print(months)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(months,sales,color='black', linestyle='solid', marker='o', markerfacecolor='black', markersize=10)
plt.title('Sales Trends')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```
